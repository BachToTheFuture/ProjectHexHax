```
import csv
```

```
df_path = "/content/drive/My Drive/NASA - Integrated Assessment/econ data/unemployment.csv"
save_path = "/content/drive/My Drive/NASA - Integrated Assessment/econ data/split/"
```

```
file = open(df_path, "r")
reader = csv.reader(file, delimiter='\t')
pc = None
writeTo = None
first = True
header = None
for log in reader:
  print(log)
  if first:
    header = log
    first = False
    continue
  country = log[0]
  if country != pc:
    print(country)
    pc = country;
    writeTo = save_path + country + ".csv"
    with open(writeTo,'a') as fd:
      fd.write('\t'.join(header) + "\n")
  with open(writeTo,'a') as fd:
    fd.write('\t'.join(log) + "\n")
```
