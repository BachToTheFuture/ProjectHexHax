```
import keras
import pandas as pd
import csv
import os
from sklearn import linear_model
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from joblib import dump, load
```

```
sp_dir = '/content/drive/My Drive/NASA - Integrated Assessment/econ data/split/'
covid_dir = '/content/drive/My Drive/NASA - Integrated Assessment/econ data/state_covid_new_cases.csv'
```

```
covid = {}
gf = open(covid_dir, "r")
reader = csv.reader(gf, delimiter=',')
first = True
for ds in reader:
  if first:
    first = False
    continue
  covid[ds[0].lower()] = ds[1:]
```

```
covX = []
UEY = []
```

```
for filename in os.listdir(sp_dir):
  cname = None
  UE = []
  gf = open(sp_dir + filename, "r")
  reader = csv.reader(gf, delimiter='\t')
  first = True
  for ds in reader:
    if first:
      first = False
      continue
    if cname == None:
      cname = ds[0]
    UE.append(ds[5])
  
  covList = covid[cname.lower()]
  UE = UE[3:]
  for k in range(len(UE)):
    covX.append([covList[k]])
    UE[k] = UE[k].split(',')
    print(UE[k])
    st = ''
    for s in UE[k]:
      st += s
    UE[k] = st
    print(UE[k])
    UEY.append([int(UE[k])])
```

```
reg = linear_model.BayesianRidge() 
reg.fit(covX,UEY)
dump(reg,"/content/drive/My Drive/NASA - Integrated Assessment/econ data/MODEL.joblib")
```
